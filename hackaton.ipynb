```python
# Python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrap_faq(url):
    # Make the GET request to the website
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all tags containing the questions
        questions = soup.find_all('h2')  # Adapt this search based on the HTML structure of the FAQ
        
        # Find all tags containing the answers
        answers = soup.find_all('p')  # Adapt this search based on the HTML structure of the FAQ
        
        # Create a list to store the questions and answers
        data = []
        
        # Add the questions and answers to the list
        for question, answer in zip(questions, answers):
            data.append({'Question': question.text.strip(), 'Answer': answer.text.strip()})
        
        # Create a DataFrame from the list of data
        df = pd.DataFrame(data)

        # Display the DataFrame
        display(df)

        # Save the DataFrame to a CSV file
        df.to_csv('faq_data.csv', index=False)
    else:
        print("HTTP request error:", response.status_code)

# URL of the FAQ to scrape
url_faq = "https://www.ikea.com/fr/fr/customer-service/knowledge/articles/dcg41290-32gb-434e-gc80-e9d869467f8b.html"  # Replace with the actual URL of the FAQ

# Call the function to scrape the FAQ and create the DataFrame

scrap_faq(url_faq)
```
